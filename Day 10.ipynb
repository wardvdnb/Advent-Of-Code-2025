### Day 10: Factory - Part 1

```python
inputs = [(l.rstrip().split(" ")) for l in open("input10.txt")]
indicators = [[False if symb=="." else True for symb in i[0][1:-1] ] for i in inputs]
buttons = [[tuple([int(x) for x in inp[1:-1].split(",")]) for inp in i[1:-1]] for i in inputs]
initial_indicators =  [[False for _ in i] for i in indicators]
joltages = [list(map(int, l)) for l in [i[-1:][0][1:-1].split(",") for i in inputs]]
```

```python
from itertools import product

def press_button(button, indicator):
    updated_indicator = [(not indicator[i]) if i in button else indicator[i] 
                         for i in range(len(indicator)) ]
    return updated_indicator

def push_buttons(goal, start, buttons, depth=0):
    if start == goal: return depth
    else:
        while(True):
            depth += 1
            for button_list in product(buttons, repeat=depth):
                indicator = start
                for button in button_list:
                    indicator = press_button(button, indicator)
                if(indicator == goal): return depth

result = 0
for i in range(len(indicators)):
    result += push_buttons(indicators[i], initial_indicators[i], buttons[i], 0)

print(result)
```

### Day 10 - Part 2

Inspired to use linprog by: https://www.reddit.com/r/adventofcode/comments/1pity70/comment/nt8vlv9/

```python
from scipy.optimize import linprog

def optimal_solution(buttons, joltages):
    opt = linprog(c=[1]*len(buttons), #A_ub=lhs_ineq, b_ub=rhs_ineq,
                A_eq=[[1 if i in button else 0 for button in buttons] for i in range(len(joltages))], 
                b_eq=joltages, #bounds=bnd,
                integrality=1) # 1 for ints, 0 for continuous variables
    return round(opt.fun)
```

```python
res = 0
for i in range(len(joltages)):
    res += optimal_solution(buttons[i], joltages[i])
print(res)
```
