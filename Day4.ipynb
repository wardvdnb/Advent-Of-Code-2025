### Day 4: Printing Department - Part 1

```python
from itertools import product

def neighbors(i, j, grid):
    neighbor_list = (
        (x, y) for (x, y) in product([i-1,i,i+1], [j-1,j,j+1]) 
        # All combinations of a 2 lists, with mirroring and duplicates 
        # (i.e. [A, B] x [A, B] = [A, A], [A, B], [B, A], [B, B])

        if (x, y) != (i, j) 
        and 0 <= x < len(grid) 
        and 0 <= y < len(grid[0])
    )
    return neighbor_list

with open("input4.txt") as file:
    """
    """
    grid = []
    for line in file:
        line = line.rstrip()
        grid.append([symb for symb in line])

    number_grid = [[0]*len(grid) for _ in range(len(grid[0]))] 
    #! DONT USE THIS: [[0]*len(grid)]*len(grid[0])
    for i in range(0, len(grid)):
        for j in range(0, len(grid[i])):
            if grid[i][j] == "@":
                for (n_i, n_j) in neighbors(i, j, grid):
                    number_grid[n_i][n_j] += 1

def part1(grid, number_grid):
    accessible = 0
    for i in range(0, len(grid)):
        for j in range(0, len(grid[i])):
            if grid[i][j] == "@":
                accessible += 1 if number_grid[i][j] < 4 else 0

    print(accessible)

part1(grid, number_grid)
```

### Day 4 - Part 2

```python
def part2(grid, number_grid):
    """
    Note: this adjusts the grids initialized by the previous exercise, so
    need to run all to re-generate a response
    """
    removed = 0
    done = False
    while(not done):
        removed_start = removed
        for i in range(0, len(grid)):
            for j in range(0, len(grid[i])):
                if grid[i][j] == "@":
                    if number_grid[i][j] < 4:
                        grid[i][j] = "."
                        removed += 1
                        for (n_i, n_j) in neighbors(i, j, grid):
                            number_grid[n_i][n_j] -= 1
        done = True if(removed_start == removed) else False
    print(removed)

part2(grid, number_grid)
```
