### Day 1: Secret Entrance

```python
password = 0
dial_value = 50
with open("input1.txt") as file:
    for line in file:
        turn = line[0]
        turn_value = int(line[1:])
        dial_value += (-turn_value if turn == "L" else turn_value) 
        dial_value %= 100
        # print("Dial is rotated ", line, " to point at ", dial_value)
        password += (1 if dial_value == 0 else 0)
        # print("Password is now: ", password)
password
```

### Day 1 - Part 2

```python
def passed_zero_n_times(old_value, new_value):
    passed = 0
    if new_value == 0: # e.g. 50 to 0
        passed = 1
    if new_value < 0:
        passed = abs(new_value // -100)
        if old_value > 0: # e.g., 50 to -10
            passed += 1
    if new_value > 0: # e.g., 10 to 340
        passed = abs(new_value // 100)
    return passed
```

```python
password = 0
dial_value = 50
with open("input1.txt") as file:
    for line in file:
        turn = line[0]
        turn_value = int(line[1:])
        old_value = dial_value
        dial_value += (-turn_value if turn == "L" else turn_value)
        password += passed_zero_n_times(old_value, dial_value)
        # print("Dial moved from ", old_value, " to ", dial_value, 
        #       f" => pointed at zero {passed_zero_n_times(old_value, dial_value)} time(s)")
        dial_value %= 100
password
```
