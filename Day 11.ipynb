### Day 11: Reactor - Part 1

```python
lines = [(l.rstrip().split(" ")) for l in open("input11.txt")]
adjacencies = {l[0][:-1]: l[1:] for l in lines}
adjacencies["out"] = set()
```

```python
#Inspired by: https://www.geeksforgeeks.org/dsa/find-paths-given-source-destination/
paths = 0

def dfsRec(adj, s, curr_path):
    curr_path.append(s)

    # Recursively visit all adjacent vertices
    for i in adj[s]:
        if i == "out":
            global paths
            paths += 1
        else:
            dfsRec(adj, i, curr_path)
    
    curr_path.pop()
        
dfsRec(adjacencies, "you", [])
print(paths)
```

### Day 11 - Part 2

Could not find this myself, had to look at the solutions, more specifically this:
https://www.reddit.com/r/adventofcode/comments/1pjp1rm/comment/ntf4e0t/
and noted that the use of @cache is very important here, as it saves states for each node
with its accompanying values for dac/fft, and simply recalls them when necessary in O(1) time (memoization?).
Alternatively could look into an explicit DP solution.

```python
from functools import cache

@cache
def dfsRec(src="svr", goal="out", dac=False, fft=False):
    if(src == "dac"):
        dac = True
    if(src == "fft"):
        fft = True
    if(src == goal and dac and fft):
        return 1

    # Recursively visit all adjacent vertices
    return sum(dfsRec(i, goal, dac, fft) for i in adjacencies[src])
        
dfsRec()
```
