### Day 9: Movie Theater - part 1

```python
red_tiles = [tuple(int(x) for x in line.strip().split(",")) for line in open("dummy-input9.txt")]

max_area = 0
for (x1, y1) in red_tiles:
    for (x2, y2) in red_tiles:
        area = max(abs(x1-x2), abs(y1-y2))+1 \
        if(x1 == x2 or y1 == y2) \
        else (abs(x1-x2)+1)*(abs(y1-y2)+1)
        
        max_area = max(max_area, area)

print(max_area)
```

### Day 9 - Part 2

```python
edges = {edge for edge in zip(red_tiles, (red_tiles[1:] + [red_tiles[0]]))}
max_pair = None
max_area = 0
for (x1, y1) in red_tiles:
    for (x2, y2) in red_tiles:
        area = 0
        if(x1 == x2 or y1 == y2):
            area = max(abs(x1-x2), abs(y1-y2))+1
        else:
            min_x, min_y, max_x, max_y = min(x1, x2), min(y1, y2), max(x1, x2), max(y1, y2)

            # ------------------------------
            # After I got stuck on pt.2 I borrowed this code to fix my solution: 
            # https://www.reddit.com/r/adventofcode/comments/1phywvn/comment/nt2m7fx
            for ((p,q), (r,s)) in edges: 
                # uses python's "For else" where the else only runs if no break occured
                if p>r: p,r = r,p
                if q>s: q,s = s,q
                if ((p < max_x and min_x < r and q < max_y and min_y < s)):
                    break
            #------------------------------
            else:
                area = (abs(x1-x2)+1)*(abs(y1-y2)+1)
                
        if(area > max_area):
            max_pair = ((x1, y1), (x2, y2))
            max_area = area

print(max_area)
print(max_pair)
```
