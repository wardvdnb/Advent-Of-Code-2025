### Day 12: Christmas Tree Farm - Part 1

```python
import re
lines = [(l.rstrip().split()) for l in open("input12.txt")]
presents = [list(map(list, present)) for present in 
            [re.split(r"\d+:", line)[1].split() 
            for line in open("dummy-input12.txt").read().split("\n\n")[:-1]]]
regions = [[tuple(int(nr) for nr in reg[0][:-1].split('x'))] + 
           [int(nr) for nr in reg[1:]] for reg in 
           [l for l in lines if re.search(r"(\d+x\d+):(.*)", "".join(l))]]
```

```python
print(
    sum(w*h >= sum([3*3*present for present in presents]) 
        for (w, h), *presents in regions))
```
