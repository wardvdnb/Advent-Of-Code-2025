### Day 5: Cafeteria - Part 1

```python
import re

with open("input5.txt") as file:
    ranges = [[int(id) for id in line.split("-")] for line in file if re.search("^.*-.*$", line.strip())]
    file.seek(0) # reset file pointer to start
    ids = [int(line.strip()) for line in file if re.search("^[0-9]+$", line.strip())]
    print(sum(1 for i in ids if any(r[0] <= i <= r[1] for r in ranges)))
```

```python
b = []
for begin,end in sorted(ranges):
    if b and b[-1][1] >= begin - 1:
        b[-1][1] = max(b[-1][1], end)
    else:
        b.append([begin, end])

print(sum([(end-start)+1 for (start, end) in b]))
```
