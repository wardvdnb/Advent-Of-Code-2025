### Day 3: Lobby - Part 1

```python
with open("input3.txt") as file:
    total = 0
    for line in file:
        line = line.rstrip()
        i1, i2 = 0, 1
        best_i1 = 0
        val1, val2 = 1, 1
        while i1 < (len(line)-1) and val1 < 9:
            new_val = int(line[i1])
            if new_val > val1:
                best_i1 = i1
                val1 = new_val
            i1 += 1
        i2 = best_i1+1
        while i2 < (len(line)) and val2 < 9:
            new_val = int(line[i2])
            if new_val > val2:
                val2 = new_val
            i2 += 1
        # print(f"{line[:best_i1]}[{val1}]{line[(best_i1+1):i2-1]}[{val2}]{line[(i2):]}")
        total += int(f"{val1}{val2}")
    print(total)
```

### Day 3: Part Two

```python
with open("input3.txt") as file:
    """
    Slide from left to right and take max value for each battery, 
    capped at 9 and making sure to leave room for the other batteries.
    (i.e. the first one leaves at least 11 batteries 
    for the others to still have a digit)
    """
    total = 0
    digit_amount = 12
    for line in file:
        line = line.rstrip()
        indexes = [0]*digit_amount
        best_i = [i for i in range(digit_amount)]
        best_vals = [0]*digit_amount
        for i in range(digit_amount):
            curr_idx = indexes[i]
            while curr_idx < (len(line)-digit_amount+i+1) and best_vals[i] < 9:
                new_val = int(line[curr_idx])
                if new_val > best_vals[i]:
                    best_i[i] = curr_idx
                    best_vals[i] = new_val
                curr_idx += 1
            if i != 11:
                indexes[i+1] = best_i[i]+1
        # print("".join([f"[{line[i]}]" if i 
        #                in best_i else f"{line[i]}" for i in range(len(line))]))
        # print("------------")
        total += int("".join([f"{val}" for val in best_vals]))
    print(total)
```
