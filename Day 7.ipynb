### Day 7: Laboratories - part 1

```python
diag = []
with open("input7.txt") as f:
    for line in f:
        diag.append(list(line.strip()))

def send_beam(start_coords, diag):
    x, y = start_coords
    if(0<= y < len(diag[0]) and (0<= x < len(diag))):
        if diag[x][y] == ".":
            diag[x][y] = "|"
            return send_beam((x+1, y), diag)
        elif diag[x][y] == "^":
            return 1 + send_beam((x, y-1), diag) + send_beam((x, y+1), diag)
    return 0
    
start_y = diag[0].index("S")
print(send_beam((1, start_y), diag))
        
```

### Day 7 - Part 2

```python
# Had to watch a visualisation of the solution to solve this one
diag = []
with open("input7.txt") as f:
    for line in f:
        if('^' in line or 'S' in line):
            diag.append([0 if symb == "." else symb for symb in list(line.strip())])

start_y = diag[0].index("S")
diag[0][start_y] = 1

for row in range(len(diag)):
    for col in range(len(diag[0])):
        if diag[row][col] == "^":
            diag[row][col-1] += diag[row-1][col]
            diag[row][col+1] += diag[row-1][col]
        elif diag[row][col] >= 0 and diag[row-1][col] != "^":
            diag[row][col] += diag[row-1][col]

print(sum([int(nr) for nr in diag[-1] if nr!="^"]))
```
