### Day 2: Gift Shop - Part 1

```python
def is_repeating_pattern(id):
    if(len(id)%2 != 0):
        return False
    else:
        reverse_first_half = id[0:len(id)//2][::-1]
        second_half = id[len(id)//2:]
        return repeated_pattern(reverse_first_half+second_half)

def repeated_pattern(substring):
    if(substring==""):
        return True
    if(len(substring)==1):
        return False
    elif(substring[0] != substring[-1]):
        return False
    else:
        return repeated_pattern(substring[1:-1])
```

```python
def is_repeating_pattern2(id):
    # Brute-force solution, alternatively use regex
    for period in (n for n in range(1, len(id)//2 + 1) if len(id)%n == 0):
        match = id[0:period]
        k = period
        while k < len(id) and (match == id[k:k+period]):
            k+=period
        if k == len(id):
            return True
        else:
            continue
    return False
```

```python
def compute_solution(part1=True):
    #part 1=False for part2
    ctr = 0
    is_repeating_pattern_func = is_repeating_pattern if part1 else is_repeating_pattern2
    with open("input2.txt") as file:
        ranges = file.readline().strip().split(",")
        for id_range in ranges:
            start_id, end_id = (int(id) for id in id_range.split("-"))
            for i in range(start_id, end_id + 1):
                id = str(i)
                if(is_repeating_pattern_func(id)):
                    ctr += int(id)
        print(ctr)
```

```python
compute_solution(part1=True)
```
