### Day 6: Trash Compactor - Part 1

```python
# From: https://stackoverflow.com/a/1740761

import operator

ops = {
    '+' : operator.add,
    '-' : operator.sub,
    '*' : operator.mul,
    '/' : operator.truediv,  # use operator.div for Python 2
    '%' : operator.mod,
    '^' : operator.xor,
}
```

```python
with open("input6.txt") as file:
    inputs = []
    for line in file:
        row = [symb for symb in line.split()] # default split by whitespace
        inputs.append(row)
    operators = inputs.pop()

    results = [0 if op == '+' else 1 for op in operators] 
    # identity for sum is 0 for product it's 1

    for i in range(len(inputs)):
        for j in range(len(inputs[0])):
            results[j] = ops[operators[j]](results[j], int(inputs[i][j]))

    print(sum(results))
```

### Day 6 - Part 2

```python
with open("input6.txt") as file:
    inputs = []
    for line in file:
        row = [id for id in line.strip("\n")]
        inputs.append(row)
    operators = "".join(inputs.pop()).split()

    numbers = []
    results = [0 if op == '+' else 1 for op in operators] 
    op_idx = 0
    for j in range(len(inputs[0])):
        # for each col build up a number string top to bottom
        number_str = ""
        for i in range(len(inputs)):
            number_str += inputs[i][j]
        if(number_str.strip(" ") == ""):
            op_idx += 1 # move on to next set of numbers
        else:
            results[op_idx] = ops[operators[op_idx]](results[op_idx], int(number_str))

    print(sum(results))
```
