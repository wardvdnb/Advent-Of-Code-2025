### Day 8: Playground - Part 1

```python
from itertools import combinations
import math

def distance(p1, p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2 + (p1[2]-p2[2])**2)
```

```python
coords = [tuple([int(value) for value in l.strip().split(",")]) for l in open("input8.txt")]
circuits = []
shortest_pairs = [] # contains for every coordinate the distance and partner coordinate
pair_distances = [(pair, distance(pair[0], pair[1])) for pair in combinations(coords, 2)]
pair_distances.sort(key=lambda x: x[1])
for _ in range(1000):
    pair, distance = pair_distances.pop(0)
    circuit1 = next((circuit for circuit in circuits if pair[0] in circuit), False)
    circuit2 = next((circuit for circuit in circuits if pair[1] in circuit), False)
    if((circuit1 and circuit2) and (circuit1 != circuit2)):
        circuits.remove(circuit1)
        circuits.remove(circuit2)
        circuits.append(circuit1.union(circuit2))
        # print("Merging circuits: ", circuit1, circuit2)
    if(circuit1 and (not circuit2)):
        circuit1.add(pair[1])
        # print("Adding coord 1 to circuit: ", pair[1])
    elif(circuit2 and (not circuit1)):
        circuit2.add(pair[0])
        # print("Adding coord 0 to circuit: ", pair[0])
    if(not any(set(pair).issubset(circuit) for circuit in circuits)):
        circuits.append(set(pair))
        # print("Making new Circuit: ", pair)

circuits.sort(key=lambda x: len(x), reverse=True)
print(math.prod((len(circuit) for circuit in circuits[:3])))
```
